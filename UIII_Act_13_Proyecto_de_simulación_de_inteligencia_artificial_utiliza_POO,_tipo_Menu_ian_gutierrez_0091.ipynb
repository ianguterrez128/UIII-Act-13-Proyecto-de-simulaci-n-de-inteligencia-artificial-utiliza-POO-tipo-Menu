### 26. Cuenta de ahorros

```python
# ======================================================
# SISTEMA DE CUENTA DE AHORROS
# CBTIS 128
# Especialidad: Inteligencia Artificial
# Submodulo 2: Elabora proyectos de programación lógica
# Docente: Eliseo Nava
# Alumno: Ian Josafat Gutierrez Saucedo
# ======================================================


# ======================================================
# CLASE Cuenta_De_Ahorros
# ======================================================

class Cuenta_De_Ahorros:

    # -----------------------------------
    # CONSTRUCTOR
    # -----------------------------------

    def __init__(self):

        self.cuentas = {}

    # ==================================================
    # ATRIBUTOS
    # ==================================================

    # nombre_cliente
    # guarda el nombre del propietario de la cuenta

    # numero_cuenta
    # almacena el numero unico de la cuenta bancaria

    # saldo
    # representa el dinero disponible en la cuenta

    # tipo_ahorro
    # indica el tipo de ahorro del cliente

    # banco
    # guarda el nombre del banco de la cuenta


    # -----------------------------------
    # FUNCION AGREGAR
    # -----------------------------------

    def agregar_cuenta(self):

        while True:

            nombre = input("\ningresa el nombre del cliente: ")
            numero = input("ingresa el numero de cuenta: ")
            saldo = float(input("ingresa el saldo inicial: "))
            tipo = input("ingresa el tipo de ahorro: ")
            banco = input("ingresa el nombre del banco: ")

            self.cuentas[nombre] = {
                "numero_cuenta": numero,
                "saldo": saldo,
                "tipo_ahorro": tipo,
                "banco": banco
            }

            print("\ncuenta agregada correctamente")

            respuesta = input("deseas agregar otra cuenta? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION LISTAR
    # -----------------------------------

    def listar_cuentas(self):

        while True:

            if len(self.cuentas) == 0:

                print("\nno hay cuentas registradas")

            else:

                print("\n================ TABLA DE CUENTAS ================\n")

                print("{:<15} {:<20} {:<15} {:<20} {:<20}".format(
                    "CLIENTE",
                    "NUMERO CUENTA",
                    "SALDO",
                    "TIPO AHORRO",
                    "BANCO"
                ))

                print("=" * 95)

                for nombre, datos in self.cuentas.items():

                    print("{:<15} {:<20} {:<15} {:<20} {:<20}".format(
                        nombre,
                        datos["numero_cuenta"],
                        datos["saldo"],
                        datos["tipo_ahorro"],
                        datos["banco"]
                    ))

                print("=" * 95)

            respuesta = input("\ndeseas volver a listar? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION DEPOSITAR
    # -----------------------------------

    def depositar_dinero(self):

        while True:

            buscar = input("\ningresa el nombre del cliente: ")

            if buscar in self.cuentas:

                deposito = float(input("ingresa la cantidad a depositar: "))

                self.cuentas[buscar]["saldo"] += deposito

                print("\ndeposito realizado correctamente")
                print("nuevo saldo:", self.cuentas[buscar]["saldo"])

            else:

                print("\ncliente no encontrado")

            respuesta = input("\ndeseas hacer otro deposito? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION RETIRAR
    # -----------------------------------

    def retirar_dinero(self):

        while True:

            buscar = input("\ningresa el nombre del cliente: ")

            if buscar in self.cuentas:

                retiro = float(input("ingresa la cantidad a retirar: "))

                if retiro <= self.cuentas[buscar]["saldo"]:

                    self.cuentas[buscar]["saldo"] -= retiro

                    print("\nretiro realizado correctamente")
                    print("saldo restante:", self.cuentas[buscar]["saldo"])

                else:

                    print("\nsaldo insuficiente")

            else:

                print("\ncliente no encontrado")

            respuesta = input("\ndeseas hacer otro retiro? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION BORRAR
    # -----------------------------------

    def borrar_cuenta(self):

        while True:

            borrar = input("\ningresa el nombre del cliente a borrar: ")

            if borrar in self.cuentas:

                del self.cuentas[borrar]

                print("\ncuenta eliminada correctamente")

            else:

                print("\ncliente no encontrado")

            respuesta = input("\ndeseas borrar otra cuenta? si/no: ")

            if respuesta == "no":
                break


# ======================================================
# OBJETO
# ======================================================

sistema = Cuenta_De_Ahorros()


# ======================================================
# MENU PRINCIPAL
# ======================================================

while True:

    print("\n================ MENU DE CUENTAS ================")
    print("1. agregar cuenta")
    print("2. listar cuentas")
    print("3. depositar dinero")
    print("4. retirar dinero")
    print("5. borrar cuenta")
    print("6. salir")

    opcion = int(input("\ningresa una opcion: "))

    if opcion == 1:

        sistema.agregar_cuenta()

    elif opcion == 2:

        sistema.listar_cuentas()

    elif opcion == 3:

        sistema.depositar_dinero()

    elif opcion == 4:

        sistema.retirar_dinero()

    elif opcion == 5:

        sistema.borrar_cuenta()

    elif opcion == 6:

        print("\nfin del programa")
        break

    else:

        print("\nopcion no valida")
```
